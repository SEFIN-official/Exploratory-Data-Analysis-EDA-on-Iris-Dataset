```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")

df.head()
```

```python
print(df.info())
print("\nSummary statistics:\n", df.describe())
print("\nMissing values:\n", df.isnull().sum())
```

```python
if 'Id' in df.columns:
    df = df.drop(columns=['Id'])
```

```python
sns.countplot(x='Species', data=df)
plt.title("Number of Samples per Species")
plt.show()
```

```python
df.hist(edgecolor='black', figsize=(10, 8))
plt.suptitle("Distribution of Features", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot of Features")
plt.show()
```

```python
sns.pairplot(df, hue="Species")
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # keep only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

```

```python
df.groupby("Species").mean()
```

##  EDA Insights from Iris Dataset

1. The dataset contains 150 samples  50 for each species: Setosa, Versicolor, Virginica.

2. Petal features (length and width) are highly useful for separating the species:
   - Setosa has the smallest petal length and width.
   - Virginica has the largest petal features.

3. Sepal features overlap between species and are less useful for classification.

4. Petal length and petal width are strongly correlated (seen in heatmap and pairplot).

5. The dataset has no missing values and no extreme outliers.

6. All species are equally represented (balanced dataset).
